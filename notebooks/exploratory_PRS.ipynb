### PRS Calculation for Autism - Michigan Imputation Server 2 Example

This data came from the Michigan Imputation Server 2. For this experimental/exploratory data analysis, I used the example dataset given by default on the server website (n = 51, European). 

*The PCA results on the imputation server suggest that the individuals sampled are of primarily northwestern European descent, with a minority of individuals clustering around southern France/northern Italy. These results will be relevant later when discussing the portability of PRS across populations.*

*The given PRS visualizations gave a reasonable estimate of the distribution of the summary statistics, but I wanted to investigate more specific measurements.*

```python
import pandas as pd 
import seaborn as sns
```

```python
data = pd.read_csv("PRS_sample/scores.txt")
```

```python
data["PGS000327"].describe()
```

```python
sns.histplot(data, x = "PGS000327", kde = True)
```

While this distribution should not be used for conclusive research purposes, since there is no phenotype data given, it could be a useful metric for comparing PRS across populations. We can see how the distribution compares to different ethnic groups, as well as cases/controls.
