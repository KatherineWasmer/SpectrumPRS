```python
import viprs as vp
import magenpy as mgp
import numpy as np
import zarr
from sklearn.model_selection import KFold
from sklearn.model_selection import StratifiedGroupKFold
import pandas as pd 
```

```python
np.random.seed(42)
```

```python
gdl = mgp.GWADataLoader(bed_files="Japan_ASD/japan_chr22/*", 
                        sumstats_files="GWAS/chr22.tsv",
                        sumstats_format="GWASCatalog",
                        ld_store_files="LD_EAS/chr_22/")
```

```python
# array of sample IDs 
sample_ids = gdl.samples 
```

https://melaniewalsh.github.io/Intro-Cultural-Analytics/02-Python/EXTRA-Randomize-Students.html

```python
# code for selecting random CV folds 
def get_folds(samples, n_folds, seed):
    np.random.seed(seed)
    np.random.shuffle(sample_ids)
    all_folds = [] 
    for i in range(n_folds):
        group = samples[i::n_folds]
        all_folds.append(group)
    return all_folds
```

```python
folds = get_folds(sample_ids, 5, 42)
```

```python
def get_cv_prs(chrom):
    np.random.seed(42)
    df_list = []
    get_folds(sample_ids, 5, 42)
    for i in range(5):
        training_folds = [item for item in sample_ids if item not in folds[i]]
        gdl_train = mgp.GWADataLoader(bed_files=f"Japan_ASD/japan_chr{chrom}/*", 
                        sumstats_files=f"GWAS/chr{chrom}.tsv",
                        sumstats_format="GWASCatalog",
                        ld_store_files=f"LD_EAS/chr_{chrom}/",
                        keep_samples=training_folds)
        gdl_test = mgp.GWADataLoader(bed_files=f"Japan_ASD/japan_chr{chrom}/*", 
                        sumstats_files=f"GWAS/chr{chrom}.tsv",
                        sumstats_format="GWASCatalog",
                        ld_store_files=f"LD_EAS/chr_{chrom}/",
                        keep_samples=folds[i])
        v = vp.VIPRS(gdl_train)
        v.fit()
        results = v.predict(gdl_test)
        df = pd.DataFrame({'ID': folds[i], f"PRS_Chr{chrom}": results})
        df_list.append(df)
    df = pd.concat(df_list)
    df_indexed = df.reset_index().drop(columns="index")
    return df_indexed 
```

```python
np.random.seed(42)
df_all = pd.DataFrame({'ID': sample_ids})
for i in range(1, 23):
    df = get_cv_prs(i)
    df_all = df_all.merge(df, on="ID")
```

```python
df_all.to_csv("Japan_prs_test2.csv", index=False)
```
